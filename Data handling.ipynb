```python
import pickle
import pandas as pd
```

# Sort the full data set by cell type and eliminate low quality cells

Full data sets are at `/data/severs/full_data_Sets/` and reduced dataset with only high quality cells are at `/data/severs/reduced_data_sets/`

```python
cell_types = ['B_cell', 'EC', 'Myeloid', 'Cancer', 'Fibroblast', 'T_cell', 'DC', 'Mast']
```

```python
full_df = pd.read_pickle("/data/severs/rna_full.pkl")
```

```python
metadata = pd.read_csv("data/rna_metadata.csv")
```

```python
cell_grouping = metadata.groupby(["CellType"])
```

```python
# Group data by celltype
cell_groups = {}
for cell_type in cell_types:
    cell_groups[cell_type] = cell_grouping.get_group(cell_type)["Cell"]
```

```python
# Splitt full dataset by cell type
datasets = {}
for cell_type in cell_types:
    datasets[cell_type] = full_df.iloc[cell_groups[cell_type].index]
```

```python
# Save full datasets for each cell type
for key in datasets:
    path = f"/data/severs/full_data_Sets/{key}.pkl"
    datasets[key].to_pickle(path)
```

```python
# Open list of high quality cells
with open("/data/severs/high_q_cell_names.pkl", 'rb') as handle:
        high_quality_cells = pickle.load(handle)
```

```python
# Select datasets with only high quality cells for each cell type
datasets_reduced = {}
for cell_type in cell_types:
    datasets_reduced[cell_type] = datasets[cell_type].loc[high_quality_cells[cell_type]]
```

```python
# Save reduced datasets that only have high quality cells
for key in datasets_reduced:
    path = f"/data/severs/reduced_data_sets/{key}.pkl"
    datasets_reduced[key].to_pickle(path)
```
